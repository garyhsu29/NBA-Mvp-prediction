```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def chunks(mylist,i, n):
    return mylist[i:i+n]
```

```python
field_name = []
team_list = []
start = 0
end = 1000
page = 11


def get_all_data(start,end):
    while start <= end: 
        
        webpage = 'https://www.basketball-reference.com/play-index/tgl_finder.cgi?request=1&player=&match=single&lg_id=NBA&year_min=1981&year_max=2018&team_id=&opp_id=&is_range=N&is_playoffs=N&round_id=&best_of=&team_seed=&opp_seed=&team_seed_cmp=eq&opp_seed_cmp=eq&game_num_type=team&game_num_min=&game_num_max=&game_month=&game_location=&game_result=&is_overtime=&c1stat=&c1comp=&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&order_by=date_game&order_by_asc=&offset='
        webpage = webpage + str(start)
        link_all_season = requests.get(webpage)
        
        if link_all_season.status_code == requests.codes.ok:
            print("OK"+ str(start) +'/' + str(end))
        else:
            print("Webpage Broke")
        
        NBA_doc = link_all_season.text
        NBA_soup = BeautifulSoup(NBA_doc, 'html.parser')
        NBA_td = NBA_soup.find_all("td")
        NBA_th = NBA_soup.find_all("th")   
        start += 100
        i = 0
        for tag in NBA_td:
            i += 1
            if i % 33 == 1 or i % 33 == 2 or i % 33 == 3 or i % 33 == 4 or i % 33 == 5:
                team_list.append(tag.string)
                
        if start == end - 100:
            for tag in NBA_th:
                field_name.append(tag.string)
            field = field_name[6:11]

    return team_list,field

my_list,my_field = get_all_data(start,end)

my_df = pd.DataFrame([my_field])
my_df.to_csv('all_team_his.csv', mode='w',index=False, header=False)
for i in range(0,len(my_list),5):
    new_list = (chunks(my_list,i,5))
    my_df = pd.DataFrame([new_list])
    my_df.to_csv('all_team_his.csv', mode='a',index=False, header=False)

```

```python
field_name = []
player_list=[]
year = 37
start = 2017

def fetch_record(NBA_tag_td,year) : 
    record = []
    i = 0
    for tag in NBA_tag_td:
        if i >= 190:
            break
        if i % 19 ==3:
            year_app = year%100
            if year_app == 0 :
                record.append(str(year-1)+'-'+"00") 
            else:
                record.append(str(year-1)+'-'+str(year_app))
        record.append(tag.string)
        i += 1
    return record

def chunks(mylist, n):
    return [mylist[i:i+n] for i in range(0, len(mylist), n)]

def get_all_mvp(start,years):
    end_year = 2017 - 37
    while start > end_year:
        webpage = 'https://www.basketball-reference.com/awards/awards_' + str (start) + '.html#mvp'
        link_mvp = requests.get(webpage)

        if link_mvp.status_code == requests.codes.ok:
            print(str(start)+" OK")
        else:
            print("Webpage Broke")


        NBA_doc = link_mvp.text
        NBA_soup = BeautifulSoup(NBA_doc, 'html.parser')
        NBA_td = NBA_soup.find_all("td")
        NBA_th = NBA_soup.find_all("th")  
        player_list.append(fetch_record(NBA_td,start))
        
        start -= 1
    k=0 
    for tag in NBA_th:
        k += 1 
        if 8 <= k <= 26:
            field_name.append(tag.string)
    return player_list

my_mvp_list = get_all_mvp(start,year)
#print(my_mvp_list)

field_name.insert(3,"Season")
my_mvp_list.insert(0,field_name)

for i in range(0,year):
    new_list = (chunks(my_mvp_list[i],20))
    my_df = pd.DataFrame(new_list)
    my_df.to_csv('all_mvp_his.csv', mode='a',index=False, header=False)
```
