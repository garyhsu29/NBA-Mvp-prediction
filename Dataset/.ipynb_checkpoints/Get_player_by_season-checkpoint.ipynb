```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
field_name = []

def fetch_record(NBA_tag_td) : 
    record = []
    for tag in NBA_tag_td:
    #i += 1 
        record.append(tag.string)
    #print(i,tag.string)
    
    return record
def chunks(mylist, n):
    return [mylist[i:i+n] for i in range(0, len(mylist), n)]
```

```python
field_name = []
player_list=[]
start = 0
end = 15797
page =158


def get_all_data(start,end):
    while start <= end:
        
        webpage = 'https://www.basketball-reference.com/play-index/psl_finder.cgi?request=1&match=single&type=totals&per_minute_base=36&per_poss_base=100&lg_id=NBA&is_playoffs=N&year_min=1947&year_max=2018&franch_id=&season_start=1&season_end=-1&age_min=0&age_max=99&shoot_hand=&height_min=0&height_max=99&birth_country_is=Y&birth_country=&birth_state=&college_id=&draft_year=&is_active=&debut_yr_nba_start=&debut_yr_nba_end=&is_hof=&is_as=&as_comp=gt&as_val=0&award=&pos_is_g=Y&pos_is_gf=Y&pos_is_f=Y&pos_is_fg=Y&pos_is_fc=Y&pos_is_c=Y&pos_is_cf=Y&qual=&c1stat=&c1comp=&c1val=&c2stat=&c2comp=&c2val=&c3stat=&c3comp=&c3val=&c4stat=&c4comp=&c4val=&c5stat=&c5comp=&c6mult=&c6stat=&order_by=season&order_by_asc=&offset='
        webpage = webpage + str(start)
        link_all_season = requests.get(webpage)
        
        if link_all_season.status_code == requests.codes.ok:
            print("OK"+ str(start) +'/' + str(end))
        else:
            print("Webpage Broke")
        
        NBA_doc = link_all_season.text
        NBA_soup = BeautifulSoup(NBA_doc, 'html.parser')
        NBA_td = NBA_soup.find_all("td")
        NBA_th = NBA_soup.find_all("th")   
        player_list.append(fetch_record(NBA_td))     
        start += 100 
    k=0 
    for tag in NBA_th:
        k += 1 
        if 6 <= k <= 36:
            field_name.append(tag.string)
    return player_list

my_list = get_all_data(start,end)
my_list.insert(0,field_name)

for i in range(0,page+1):
    new_list = (chunks(my_list[i],31))
    my_df = pd.DataFrame(new_list)
    my_df.to_csv('all_his.csv', mode='a',index=False, header=False)
```
